```python
import pandas as pd


# Step 1: Read the CSV file into a pandas DataFrame


# Step 2: Assume the CSV has columns 'x', 'y', and 'class'
# Example: df = pd.read_csv('your_file.csv')
# Ensure that you have the relevant columns 'x', 'y', and 'class'

# Step 3: Define the marker shapes for each class
markers = { 0: 'circle', 1: 'triangle-up', 2: 'diamond'}

# Step 4: Create the figure and add scatter traces based on the class
fig = go.Figure()

for cls in df['DEATH_EVENT'].unique():
    class_df = df[df['DEATH_EVENT'] == cls] #In this line, we are using Boolean indexing to filter the original DataFrame df based on the value of cls.

    fig.add_trace(go.Scatter(
        x=class_df['age'],
        y=class_df['creatinine_phosphokinase'],
        mode='markers',
        marker=dict(symbol = markers[cls] ),   #symbol=markers.get(cls)),  # Use a default if the class is not found
        name=f'DEATH_EVENT{cls}'  # Legend label
    ))
# Add axis labels
fig.update_layout(
    xaxis_title='Age',  # Label for the x-axis
    yaxis_title='Creatinine Phosphokinase',  # Label for the y-axis
)

# Step 5: Show the plot
fig.show()
```

```python
markers = {0: 'circle', 1: 'cross'}  # Assign a shape to each class

# Step 3: Create a 3D scatter plot using Graph Objects
fig = go.Figure()

# Step 4: Loop through the unique classes and add them to the plot
for cls in df['DEATH_EVENT'].unique():
    class_df = df[df['DEATH_EVENT'] == cls]

    fig.add_trace(go.Scatter3d(
        x=class_df['age'],
        y=class_df['creatinine_phosphokinase'],
        z=class_df['ejection_fraction'],  # Assuming the DataFrame has a 'z' column for the third axis
        mode='markers',
        marker=dict(symbol=markers[cls], size=6),  # Use the marker shape for each class
        name=f'Class {cls}'  # Legend label for each class
    ))
# Step 5: Add axis labels
fig.update_layout(
    scene=dict(
        xaxis_title='Age',  # Label for the x-axis
        yaxis_title='Creatinine Phosphokinase',  # Label for the y-axis
        zaxis_title='Ejection Fraction'  # Label for the z-axis
    ),
)
fig.show()
```

```python
markers = {0: 'circle', 1: 'square'}  # Assign a shape to each class

# Step 3: Create a scatter plot using Graph Objects
fig = go.Figure()

# Step 4: Loop through the unique classes and add them to the plot
for cls in df['DEATH_EVENT'].unique():
    class_df = df[df['DEATH_EVENT'] == cls]

    # Add scatter plot for the class
    fig.add_trace(go.Scatter(
        x=class_df['age'],
        y=class_df['creatinine_phosphokinase'],
        mode='markers',
        marker=dict(symbol=markers[cls], size=6),  # Use the marker shape for each class
        name=f'Class {cls}'  # Legend label for each class
    ))

    # Step 5: Calculate the mean for each class
    mean_x = class_df['age'].mean()
    mean_y = class_df['creatinine_phosphokinase'].mean()

    # Add mean points with a larger marker size
    fig.add_trace(go.Scatter(
        x=[mean_x],
        y=[mean_y],
        mode='markers',
        marker=dict(symbol=markers[cls], size=12, color='red'),  # Same shape, larger size, red color
        name=f'Mean Class {cls}'  # Legend label for mean points
    ))
#  Add axis labels
fig.update_layout(
    xaxis_title='Age',  # Label for the x-axis
    yaxis_title='Creatinine Phosphokinase',  # Label for the y-axis
)

fig.show()
```

```python
# Step 3: Create a subplot figure for the histograms
fig = make_subplots(
    rows=1, cols=3,  # One row, three columns for each dimension
    subplot_titles=('Age Histogram', 'Creatinine Phosphokinase Histogram', 'Ejection Fraction Histogram')
)

# Step 4: Plot histogram for 'age'
fig.add_trace(go.Histogram(
    x=df['age'],
    nbinsx=20,  # Set number of bins
    name='Age'
), row=1, col=1)

# Step 5: Plot histogram for 'creatinine_phosphokinase'
fig.add_trace(go.Histogram(
    x=df['creatinine_phosphokinase'],
    nbinsx=20,  # Set number of bins
    name='Creatinine Phosphokinase'
), row=1, col=2)

# Step 6: Plot histogram for 'ejection_fraction'
fig.add_trace(go.Histogram(
    x=df['ejection_fraction'],
    nbinsx=20,  # Set number of bins
    name='Ejection Fraction'
), row=1, col=3)

# Step 7: Update layout to add axis titles and adjust spacing
fig.update_layout(
    title_text="Histograms of First Three Dimensions",
    showlegend=False,
    height=400,
    width=1200,
    bargap=0.1,  # Adjust gap between bars
    xaxis_title='Value',  # For the first subplot
    yaxis_title='Frequency',  # For the first subplot
    xaxis2_title='Value',  # For the second subplot
    yaxis2_title='Frequency',  # For the second subplot
    xaxis3_title='Value',  # For the third subplot
    yaxis3_title='Frequency',  # For the third subplot
)

# Step 8: Show the plot
fig.show()
```

```python
age_std = df['age'].std()
creatinine_phosphokinase_std = df['creatinine_phosphokinase'].std()
ejection_fraction_std = df['ejection_fraction'].std()

# Step 3: Print the results
print(f"Standard Deviation of Age: {age_std}")
print(f"Standard Deviation of Creatinine Phosphokinase: {creatinine_phosphokinase_std}")
print(f"Standard Deviation of Ejection Fraction: {ejection_fraction_std}")
```

```python
fig = go.Figure()

# Add box plot for 'age'
fig.add_trace(go.Box(
    y=df['age'],
    name='Age',
    boxmean=True  # Show mean line inside the box plot
))

# Add box plot for 'creatinine_phosphokinase'
fig.add_trace(go.Box(
    y=df['creatinine_phosphokinase'],
    name='Creatinine Phosphokinase',
    boxmean=True  # Show mean line inside the box plot
))

# Add box plot for 'ejection_fraction'
fig.add_trace(go.Box(
    y=df['ejection_fraction'],
    name='Ejection Fraction',
    boxmean=True  # Show mean line inside the box plot
))

# Step 3: Update layout to add axis titles
fig.update_layout(
    title="Box Plots for First Three Dimensions",
    xaxis_title="Dimension",
    yaxis_title="Values",
    showlegend=False
)

# Step 4: Show the plot
fig.show()
```

```python
classes = df['DEATH_EVENT'].unique() #for DEATH_EVENT=0 the class is categorized as "Survived", otherwise "Died"

# Create a box plot for each class
for cls in classes:
    class_df = df[df['DEATH_EVENT'] == cls]

    # Create a new figure for each class
    fig = go.Figure()

    # Add box plot for 'age'
    fig.add_trace(go.Box(
        y=class_df['age'],
        name='Age',
        boxmean=True  # Show mean line inside the box plot
    ))

    # Add box plot for 'creatinine_phosphokinase'
    fig.add_trace(go.Box(
        y=class_df['creatinine_phosphokinase'],
        name='Creatinine Phosphokinase',
        boxmean=True  # Show mean line inside the box plot
    ))

    # Add box plot for 'ejection_fraction'
    fig.add_trace(go.Box(
        y=class_df['ejection_fraction'],
        name='Ejection Fraction',
        boxmean=True  # Show mean line inside the box plot
    ))
    #Update layout to add axis titles
    if cls == 0:
        fig.update_layout(
            title=f"Box Plots for Survived",
            xaxis_title="Dimension",
            yaxis_title="Values",
            showlegend=False
        )
    else:
        fig.update_layout(
            title=f"Box Plots for Died",
            xaxis_title="Dimension",
            yaxis_title="Values",
            showlegend=False
        )



    # Step 5: Show the plot for each class
    fig.show()
```

```python
#  Select the relevant columns for the scatter matrix
dimensions = ['age', 'creatinine_phosphokinase', 'ejection_fraction']

#  Create the scatter matrix
fig = px.scatter_matrix(
    df,
    dimensions=dimensions,
    color="DEATH_EVENT",  # Color points by class (optional)
    title="3x3 Scatter Matrix of First Three Dimensions",
)

# Show the plot
fig.show()
```

```python
dimensions = ['age', 'creatinine_phosphokinase', 'ejection_fraction']

# Step 3: Create the parallel coordinates plot
fig = px.parallel_coordinates(
    df,
    dimensions=dimensions,
    color="DEATH_EVENT",  # Optional: Color by class (can be omitted if not needed)
    title="Parallel Coordinates for First Three Dimensions",
    labels={
        'age': 'Age',
        'creatinine_phosphokinase': 'Creatinine Phosphokinase',
        'ejection_fraction': 'Ejection Fraction',
        'DEATH_EVENT': 'Class'  # Label for the color dimension
    },
    color_continuous_scale='Viridis',  # Optional: color scale for better visualization
)

# Step 4: Show the plot
fig.show()
```
